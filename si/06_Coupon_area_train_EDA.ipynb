### 3. SMALL_AREA_NAME

```python
plt.figure(figsize=(14, 12))
sns.countplot(y = coupon_area['SMALL_AREA_NAME'], data = coupon_area, order = coupon_area['SMALL_AREA_NAME'].value_counts().index)
plt.show()
```

### 4. PREF_NAME 

```python
plt.figure(figsize=(14, 12))
sns.countplot(y = coupon_area['PREF_NAME'], data = coupon_area, order = coupon_area['PREF_NAME'].value_counts().index)
plt.show()
```

### 5. COUPON_ID_hash

```python
COUPON_ID = coupon_area.groupby("COUPON_ID_hash").size().reset_index(name='counts')
COUPON_ID = COUPON_ID.sort_values(by=['counts'], ascending=False)
COUPON_ID = COUPON_ID.reset_index(drop=True)
COUPON_ID.head(10)
```
